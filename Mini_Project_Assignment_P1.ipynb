```python
# 3.6 Replacing values
part1_df = df.copy().reset_index(drop=True)

country_replacements = {
    'EIRE': 'Ireland',
    'RSA': 'South Africa',
    'Unspecified': np.nan
}
part1_df['Country'] = part1_df['Country'].replace(country_replacements)
part1_df['Country'].value_counts(dropna=False)
```

```python
# 3.7 Renaming columns
rename_map = {
    'InvoiceNo': 'invoice_no',
    'StockCode': 'stock_code',
    'Description': 'description',
    'Quantity': 'quantity',
    'InvoiceDate': 'invoice_date',
    'UnitPrice': 'unit_price',
    'CustomerID': 'customer_id',
    'Country': 'country',
    'LineRevenue': 'line_revenue'
}
part1_df = part1_df.rename(columns=rename_map)
part1_df.head()
```

```python
# 3.10 Handling missing values
missing_summary = pd.DataFrame({
    'missing_count': part1_df.isnull().sum(),
    'missing_percent': (part1_df.isnull().mean() * 100).round(2)
}).query('missing_count > 0')

customer_id_dropna_demo = part1_df.dropna(subset=['customer_id'])
description_dropna_demo = part1_df.dropna(subset=['description'])
description_fillna_demo = part1_df['description'].fillna('Unknown Product')

part1_df['customer_id_missing'] = part1_df['customer_id'].isnull()
part1_df['description'] = part1_df['description'].fillna('Unknown Product')

missing_policy = pd.DataFrame({
    'column': ['customer_id', 'description'],
    'policy': [
        'keep missing customer_id for sales analysis; exclude only for customer-level summaries',
        'fill missing description for profiling, then remove unknown product rows from final sales data'
    ]
})

missing_summary
```

```python
# 3.11 Deleting a column
columns_to_drop = [col for col in ['line_revenue'] if col in part1_df.columns]
dropped_column_log = pd.DataFrame({
    'dropped_column': columns_to_drop,
    'reason': ['recreated as revenue after final cleaning'] * len(columns_to_drop)
})
part1_df = part1_df.drop(columns=columns_to_drop, errors='ignore')
dropped_column_log
```

```python
# 3.12 Deleting a row
non_product_codes = {
    'POST', 'DOT', 'M', 'BANK CHARGES', 'AMAZONFEE',
    'D', 'C2', 'S', 'PADS', 'CRUK'
}

is_cancelled = part1_df['invoice_no'].astype(str).str.startswith('C', na=False)
bad_quantity = part1_df['quantity'] <= 0
bad_price = part1_df['unit_price'] <= 0
unknown_description = part1_df['description'].astype(str).str.strip().eq('Unknown Product')
blank_description = part1_df['description'].astype(str).str.strip().eq('')
non_product_line = part1_df['stock_code'].astype(str).str.upper().isin(non_product_codes)

remove_rows = (
    is_cancelled |
    bad_quantity |
    bad_price |
    unknown_description |
    blank_description |
    non_product_line
)

sales_df = part1_df.loc[~remove_rows].copy()

row_removal_summary = pd.Series({
    'start_rows': len(part1_df),
    'cancelled_invoice_rows': int(is_cancelled.sum()),
    'quantity_less_equal_zero_rows': int(bad_quantity.sum()),
    'unit_price_less_equal_zero_rows': int(bad_price.sum()),
    'unknown_or_blank_description_rows': int((unknown_description | blank_description).sum()),
    'non_product_rows': int(non_product_line.sum()),
    'rows_removed_by_filter': int(remove_rows.sum()),
    'sales_rows_before_duplicates': len(sales_df)
})

row_removal_summary
```

```python
# 3.13 Dropping duplicates
exact_duplicate_count = sales_df.duplicated().sum()
clean_sales = sales_df.drop_duplicates().copy()
rows_dropped_duplicates = len(sales_df) - len(clean_sales)

pd.Series({
    'exact_duplicates_before_drop': int(exact_duplicate_count),
    'rows_dropped_duplicates': int(rows_dropped_duplicates),
    'clean_sales_rows': len(clean_sales)
})
```

```python
# 3.18 Applying a function
clean_sales['revenue'] = clean_sales['quantity'] * clean_sales['unit_price']
clean_sales['description'] = clean_sales['description'].apply(lambda x: str(x).strip().title())
clean_sales['is_cancelled'] = clean_sales['invoice_no'].apply(lambda x: str(x).startswith('C'))

clean_sales[['invoice_no', 'stock_code', 'description', 'quantity', 'unit_price', 'revenue', 'is_cancelled']].head()
```

```python
# 3.17 Looping over a column
sample_descriptions = clean_sales['description'].head(10)

loop_cleaned_descriptions = []
for item in sample_descriptions:
    loop_cleaned_descriptions.append(str(item).strip().title())

list_comprehension_flags = [str(item).startswith('C') for item in part1_df['invoice_no'].head(10)]
apply_flags = part1_df['invoice_no'].head(10).apply(lambda x: str(x).startswith('C')).tolist()

looping_comparison = pd.DataFrame({
    'invoice_no': part1_df['invoice_no'].head(10).tolist(),
    'list_comprehension_flag': list_comprehension_flags,
    'apply_flag': apply_flags
})

looping_note = 'apply keeps the operation inside the pandas workflow; vectorized string methods are best when available.'
looping_comparison
```

```python
# 3.14 Grouping by values
revenue_by_country = clean_sales.groupby('country')['revenue'].sum().sort_values(ascending=False)
orders_per_customer = (
    clean_sales.dropna(subset=['customer_id'])
    .groupby('customer_id')['invoice_no']
    .nunique()
    .sort_values(ascending=False)
)

revenue_by_country.head(10)
```

```python
# 3.16 Aggregating
invoice_totals = clean_sales.groupby(['country', 'invoice_no'], as_index=False)['revenue'].sum()
country_agg = (
    invoice_totals.groupby('country')
    .agg(
        total_revenue=('revenue', 'sum'),
        mean_order_value=('revenue', 'mean'),
        transaction_count=('invoice_no', 'nunique')
    )
    .sort_values('total_revenue', ascending=False)
)

customer_invoice_totals = (
    clean_sales.dropna(subset=['customer_id'])
    .groupby(['customer_id', 'invoice_no'], as_index=False)['revenue']
    .sum()
)
customer_agg = (
    customer_invoice_totals.groupby('customer_id')
    .agg(
        total_revenue=('revenue', 'sum'),
        mean_order_value=('revenue', 'mean'),
        transaction_count=('invoice_no', 'nunique')
    )
    .sort_values('total_revenue', ascending=False)
)

country_agg.head(10)
```

```python
# 3.19 Applying to groups
def summarize_customer(group):
    dates = pd.to_datetime(group['invoice_date'], errors='coerce')
    return pd.Series({
        'total_spend': group['revenue'].sum(),
        'orders': group['invoice_no'].nunique(),
        'active_months': dates.dt.to_period('M').nunique()
    })

customer_custom_summary = (
    clean_sales.dropna(subset=['customer_id'])
    .groupby('customer_id')
    .apply(summarize_customer, include_groups=False)
    .sort_values('total_spend', ascending=False)
)

customer_custom_summary.head(10)
```

```python
# 3.15 Grouping by time
import matplotlib.pyplot as plt

clean_sales['invoice_date'] = pd.to_datetime(clean_sales['invoice_date'], errors='coerce')
clean_sales = clean_sales.dropna(subset=['invoice_date']).copy()

clean_sales_time_indexed = clean_sales.set_index('invoice_date').sort_index()
monthly_revenue = clean_sales_time_indexed['revenue'].resample('ME').sum()
weekly_revenue = clean_sales_time_indexed['revenue'].resample('W').sum()

monthly_revenue.plot(kind='line', title='Monthly Revenue')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.tight_layout()
plt.show()

monthly_revenue
```

```python
# 3.20 Concatenating
first_half_sales = clean_sales[clean_sales['invoice_date'] < '2011-07-01'].copy()
second_half_sales = clean_sales[clean_sales['invoice_date'] >= '2011-07-01'].copy()

recombined_sales = pd.concat([first_half_sales, second_half_sales], ignore_index=True)

pd.Series({
    'first_half_rows': len(first_half_sales),
    'second_half_rows': len(second_half_sales),
    'recombined_rows': len(recombined_sales),
    'clean_sales_rows': len(clean_sales)
})
```
